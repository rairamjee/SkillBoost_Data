## Imports

```python

import pandas as pd
import os 
import matplotlib.pyplot as plt
import seaborn as sns
x = "ch:start=.2,rot=-.3"
```

```python
os.makedirs('04Report',exist_ok=True)
```

##### Data Load

```python
users_report_df=pd.DataFrame(pd.read_csv('03Transform/usersTrans.csv'))
domain_details_report_df=pd.DataFrame(pd.read_csv('03Transform/domainDetailsTrans.csv'))
organisation_review_report_df=pd.DataFrame(pd.read_csv('03Transform/organisationReviewTrans.csv'))
trainings_avg_success_report_df=pd.DataFrame(pd.read_csv('03Transform/trainingAvgSuccessRateTrans.csv'))
training_overview_report_df=pd.DataFrame(pd.read_csv('03Transform/trainingOverviewsTrans.csv'))
```

##### Users 

```python

designation_counts = users_report_df['designation'].value_counts().reset_index()
designation_counts.columns = ['designation', 'count']  

plt.figure(figsize=(15, 6))
sns.barplot(data=designation_counts, x='designation', y='count',palette = "ch:start=.2,rot=-.3")
plt.title('Count of Employees by Designation')
plt.xlabel('Designation')
plt.ylabel('Number of Employees')
# plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
role_count=users_report_df['role'].value_counts().reset_index()
role_count.columns=['roles','count']

plt.figure(figsize=(4, 4))
sns.barplot(data=role_count,x='roles',y='count',palette = "ch:start=.2,rot=-.3")
plt.title("Role Distributon")
plt.xlabel("Roles")
plt.ylabel("Count")
```

##### Training

```python

highest_scores = training_overview_report_df[['trainingId', 'highest_score', 'total_responses']]
print("Columns before renaming:", highest_scores.columns)

plt.figure(figsize=(30,6))

sns.barplot(x='trainingId', y='total_responses', data=highest_scores, palette = "ch:start=.2,rot=-.3", alpha=0.6, label='Total Responses')
plt.title('Responses in Each trainings ')
plt.xlabel('Training ID')
plt.ylabel('Total Responses')
plt.tight_layout()  
plt.show()


```

```python
plt.figure(figsize=(30,6))

sns.lineplot(x='trainingId', y='highest_score', data=highest_scores, color='green', alpha=0.6, label='Scores',marker='o')
plt.title('Highest Marks in Each trainings ')
plt.xlabel('Training ID')
plt.ylabel('Highest Marks')
# plt.tight_layout()  

plt.show()
```

```python
palette = 'viridis'
plt.figure(figsize=(25, 6))
sns.barplot(data=trainings_avg_success_report_df, x='trainingId', y='avg_scores', palette = "ch:start=.2,rot=-.3", alpha=0.6)
# sns.lineplot(data=trainings_avg_success_report_df, x='trainingId', y='success_rate_percentage', color='red', label='Success Rate', marker='o')
plt.title('Average Scores by Training ID')
plt.xlabel('Training ID')
plt.ylabel('Scores')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
palette = 'viridis'
plt.figure(figsize=(25, 6))
sns.lineplot(data=trainings_avg_success_report_df, x='trainingId', y='success_rate_percentage', color='green', label='Success Rate', marker='o')
plt.title(' Success Rates by Training ID')
plt.xlabel('Training ID')
plt.ylabel('Success Rate (%)')
plt.legend()
plt.tight_layout()
plt.show()
```

##### Domain

```python
domain_df=domain_details_report_df[['domainName','training_count']]
plt.figure(figsize=(10,5))
sns.barplot(data=domain_df,x='domainName',y='training_count',palette="ch:start=.2,rot=-.3")
plt.title('No of Trainings in Each Domain')
plt.xlabel('Domain Name')
plt.ylabel('Training Count')
plt.show()
```

##### Organisation Review 

```python
avg_performance_df = organisation_review_report_df.groupby('userId')['performanceScore'].mean().reset_index()
avg_performance_df.columns = ['userId', 'avg_performanceScore']

plt.figure(figsize=(25, 5))
sns.scatterplot(data=avg_performance_df, x='userId', y='avg_performanceScore',color='green')
plt.axhline(y=1.3, color='red', linestyle='--', label='Threshold (1.5)')
plt.title('Average Performance Score by User ID')
plt.xlabel('User ID')
plt.ylabel('Average Performance Score')

plt.tight_layout()
plt.show()
```

##### Data Load in Report Layer

```python
users_report_df.to_csv('04Report/usersReport.csv',index=False)
domain_details_report_df.to_csv('04Report/domainDetailsReport.csv',index=False)
organisation_review_report_df.to_csv('04Report/organisationReviewReport.csv',index=False)
training_overview_report_df.to_csv('04Report/trainingOverviewsReport.csv',index=False)
trainings_avg_success_report_df.to_csv('04Report/trainingAvgSuccessRateReport.csv',index=False)
```
